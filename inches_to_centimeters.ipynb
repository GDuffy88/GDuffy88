```python
import pandas as pd

import numpy as np
data=pd.read_csv('/content/inch_cm-1 (1).csv')
print(data.head())
print(data.describe())
```

```python
x=data[['inch']].to_numpy()
y=data[['cm']].to_numpy()
```

```python


seventy_percent = int(len(x) * 0.7)
x_train = x[:seventy_percent]
y_train = y[:seventy_percent]

x_test = x[seventy_percent:]
y_test = y[seventy_percent:]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

Linear Regression: Making predictions with a model i.e.,training an algorithm

```python
#we get the y_pred by using the generated model amd the x_test data
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error as m_s_e

# Sample Actual values from the data
DataTrue =[3,0.0,2,7.2]

#Predicted values using a Data Science Model
DataPredict =[2.8, -0.1, 2, 7.4]

#Calculate RMSE
mse = m_s_e(DataTrue, DataPredict)
rmse=np.sqrt(mse)
print("Root Mean Squared Error:", mse)
```
